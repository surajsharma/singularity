```python
import scrapy
from scrapy.crawler import CrawlerProcess

class scraper(scrapy.Spider):
    name = 'channeldldr'
    allowed_domains = ['youtube.com']

    start_urls = ['https://www.youtube.com/@ansuzsociety/videos']

    def parse(self, response):
        for cont in response.xpath('//div[@id="meta"]'):
            print("HERE")
            vid_link = response.xpath('//a[@id="video-title-link"]').get()
            print('LINK', vid_link)
            yield{'link' : vid_link}

process = CrawlerProcess(settings = {'OP' : {'op.csv' : {'format':'csv'}}})
process.crawl(scraper)
process.start()
```
