```python
from pydantic_ai import Agent
import nest_asyncio
import asyncio
import os
from google.colab import userdata

# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()

os.environ['GEMINI_API_KEY'] = userdata.get('GEMINI_API_KEY')

agent = Agent(
    'gemini-1.5-flash',
    system_prompt='Be concise, reply with one sentence.',
)

# Use asyncio.run to execute the coroutine within the nested loop
result = asyncio.run(agent.run('act as a life coach giving motivational advice'))

print(result.data)
```
